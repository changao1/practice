```python
def generate_data(n, generator_type):
    ϵ_values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0,1)
        else:
            e = np.random.randn()
        ϵ_values.append(e)
    return ϵ_values
data = generate_data(100, 'U')
plt.plot(data)
plt.show()
        
```

```python
def factorial(n):
    N = 1
    N = N*n
    while n > 1:
        n -= 1
        N = N*n
    return N
#not right when n = 0

# def factorial(n):
#     k = 1
#     for i in range(n):
#         k = k*(i+1)
#     return k
```

```python
# pay if k successes in a sequence
def draw_new(k):
    payoff = 0
    count = 0
    
    for i in range(10):
        U = uniform()
        count = count + (1 if U < 0.5 else 0)
        print(count)
        if count == k:
            payoff = 1
    return payoff
```
