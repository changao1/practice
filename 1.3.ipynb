```python
r = 0.025
T = 50
b = np.empty(T+1)
b[0] = 10

for t in range(T):
    b[t+1] = (1+r)*b[t]

plt.plot(b, label = 'bank balance')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    x[t+1] = alpha*x[t] + np.random.randn()

plt.plot(x, label = 'Time series')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
T = 200
x = np.empty(T+1)
x[0] = 0
for alpha in [0, 0.8, 0.98]:   
    for t in range(T):
        x[t+1] = alpha*x[t] + np.random.randn()
    plt.plot(x, label = f'$\\alpha={alpha}$')
    
    
plt.legend()
plt.show()
```

```python

alpha = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    x[t+1] = alpha*np.abs(x[t]) + np.random.randn()

plt.plot(x, label = 'Time series')
plt.legend()
plt.show()
```
